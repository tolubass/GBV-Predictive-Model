```python
def determine_vulnerability(row):
    score = 0
    age = row['Age of survivor']
    employment = str(row['SURVIVOR/VICTIM EMPLOYMENT STATUS']).strip().lower()
    marital_status = str(row['MARITAL STATUS']).strip().lower()
    violence_type = str(row['TYPE OF VIOLENCE']).strip().lower()
    living_with = str(row['WHO SURVIVOR/VICTIM LIVE WITH']).strip().lower()
    vulnerability_category = str(row['VULNERABLE POPULATION']).strip().lower()

    # Scoring  
    if age < 18 or age > 65:  
        score += 2  
    if employment in ['unemployed', 'informal sector']:  
        score += 1  
    if marital_status in ['widowed', 'separated', 'divorced']:  
        score += 1  
    if violence_type in ['sexual violence', 'emotional/psychological abuse']:  
        score += 2  
    if living_with in ['lives alone', 'lives with abuser']:  
        score += 2  
    if vulnerability_category not in ['no vulnerability', 'unknown', '', 'none', 'nan']:  
        score += 2  

    # Threshold (you can adjust this)  
    return 'VULNERABLE' if score >= 3 else 'NOT_VULNERABLE'
```

```python
df1['VULNERABILITY_STATUS'] = df1.apply(determine_vulnerability, axis=1)
```

```python
import pandas as pd

def determine_vulnerability_safe(row, threshold=3):
    score = 0
    reasons = []

    # safe reading + normalization
    raw_age = row.get('Age of survivor', None)
    try:
        age = float(raw_age) if raw_age not in (None, '', float('nan')) else None
    except (ValueError, TypeError):
        age = None

    def norm(x):
        return '' if pd.isna(x) else str(x).strip().lower()

    employment = norm(row.get('SURVIVOR/VICTIM EMPLOYMENT STATUS'))
    marital_status = norm(row.get('MARITAL STATUS'))
    violence_type = norm(row.get('TYPE OF VIOLENCE'))
    living_with = norm(row.get('WHO SURVIVOR/VICTIM LIVE WITH'))
    vulnerability_category = norm(row.get('VULNERABLE POPULATION'))

    # sets with possible synonyms
    unemp_set = {'unemployed', 'informal sector', 'informal', 'not employed'}
    high_risk_violence = {'sexual violence', 'emotional/psychological abuse', 'emotional abuse', 'psychological abuse'}
    living_risk = {'lives alone', 'alone', 'lives with abuser', 'with abuser', 'living with abuser'}

    if age is not None and (age < 18 or age > 65):
        score += 2
        reasons.append('age')

    if employment in unemp_set:
        score += 1
        reasons.append('employment')

    if marital_status in {'widowed', 'separated', 'divorced'}:
        score += 1
        reasons.append('marital_status')

    if violence_type in high_risk_violence:
        score += 2
        reasons.append('violence_type')

    if living_with in living_risk:
        score += 2
        reasons.append('living_with')

    if vulnerability_category not in {'', 'no vulnerability', 'unknown', 'none'}:
        score += 2
        reasons.append('vulnerable_category')

    label = 'VULNERABLE' if score >= threshold else 'NOT_VULNERABLE'
    return {'label': label, 'score': score, 'reasons': reasons}
```

```python
# Apply the function to each row
results = df2.apply(lambda row: determine_vulnerability_safe(row), axis=1)

# If you want to split the dictionary into separate columns
results_df = pd.DataFrame(results.tolist())

# Join back to your original df2
df2 = pd.concat([df2, results_df], axis=1)

# Now df2 will have extra columns: label, score, reasons
print(df2.head())
```
